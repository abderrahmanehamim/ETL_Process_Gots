```python

def extract_company_details_xpath(driver):
    company_details = {}

    xpath_expressions = {
        'Company': "//*[@id='xFormH1-0']",
        'Country': "//*[@id='xFormText-3']",
        'Field of operation': "//*[@id='xFormText-5']",
        'Product category': "//*[@id='xFormText-9']",
        'Contact name': ["//*[@id='xFormText-12']", "//*[@id='xFormText-13']"], 
        'Email address': "//*[@id='xFormText-19']/a",
        'Address': "//*[@id='xFormText-21']",
        'License number': "//*[@id='xFormText-38']",
        'Product details': "//*[@id='xFormText-43']",
        'Certification Body': "//*[@id='xFormText-45']",
        'Certificate Expiry Date': "//*[@id='xFormText-49']"
    }

    for field, xpaths in xpath_expressions.items():
        if isinstance(xpaths, list):
            # Check if it's a list of XPaths
            for xpath in xpaths:
                try:
                    element = WebDriverWait(driver, 5).until(
                        EC.presence_of_element_located((By.XPATH, xpath))
                    )
                    company_details[field] = element.text.strip()
                    break 
                    # Found the element, stop checking other XPaths
                except (TimeoutException, NoSuchElementException):
                    continue 
                    # Try the next XPath if this one fails
        else: 
            # If it's a single XPath, handle as before
            try:
                element = WebDriverWait(driver, 5).until(
                    EC.presence_of_element_located((By.XPATH, xpaths))
                )
                company_details[field] = element.text.strip()
            except (TimeoutException, NoSuchElementException):
                company_details[field] = "N/A"
    return company_details
```
